# Gender Detection
#### only for frontal face

```python
import cv2
path = input("path :")
img = cv2.imread(path)
#shows the model accuracy
MODEL_MEAN_VALUES = (78.4263377603, 87.7689143744, 114.895847746) # this values are fixed

gender_list = ['Male', 'Female']

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
faceDetect= cv2.CascadeClassifier("E:\Training Ducat\documents of ml\haarcascade_frontalface_default.xml")

faces = faceDetect.detectMultiScale(gray,1.2,7)
for (x,y,w,h) in faces:
    cv2.rectangle(img, (x,y), (x+w, y+h), (0,0,200), 2)

    #get face
    face_img = img[y:y+h, h:h+w].copy()
    blob  = cv2.dnn.blobFromImage(face_img,1,(227,227), 
                                 MODEL_MEAN_VALUES, swapRB= False)
#     print(blob)
    
    gender_net = cv2.dnn.readNetFromCaffe('E:\Training Ducat\documents of ml\deploy_gender.prototxt',
                                         'E:\Training Ducat\documents of ml\gender_net.caffemodel')
    
    print(gender_net)
    #Predict Gender
    gender_net.setInput(blob)
    gender_preds = gender_net.forward()
    print(gender_preds)
    gender = gender_list[gender_preds[0].argmax()]
    print(gender)
    
    cv2.putText(img, gender, (x,y-5), cv2.FONT_HERSHEY_PLAIN,2,(0,0,20),2)
    
cv2.imshow("Gender", img)
cv2.waitKey(0)
cv2.destroyAllWindow()
    
```

# for live image


```python
import cv2
cam = cv2.VideoCapture(0)
#shows the model accuracy
MODEL_MEAN_VALUES = (78.4263377603, 87.7689143744, 114.895847746) # this values are fixed

gender_list = ['Male', 'Female']
faceDetect= cv2.CascadeClassifier("C:\\Users\\pkmis\\Documents\\Machine learning\\Detection OpenCV\\Face Eye And Gender\\data\\haarcascade_frontalface_default.xml")

while True:
    ret, img = cam.read()
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    faces = faceDetect.detectMultiScale(gray,1.2,7)
    for (x,y,w,h) in faces:
        cv2.rectangle(img, (x,y), (x+w, y+h), (0,0,200), 2)

        #get face
        face_img = img[y:y+h, h:h+w].copy()
        blob  = cv2.dnn.blobFromImage(face_img,1,(227,227), 
                                 MODEL_MEAN_VALUES, swapRB= False)
#     print(blob)
    
        gender_net = cv2.dnn.readNetFromCaffe('C:\\Users\\pkmis\\Documents\\Machine learning\\Detection OpenCV\\Face Eye And Gender\\data\\deploy_gender.prototxt',
                                         'C:\\Users\\pkmis\\Documents\\Machine learning\\Detection OpenCV\\Face Eye And Gender\\data\\gender_net.caffemodel')
    
        print(gender_net)
    #Predict Gender
        gender_net.setInput(blob)
        gender_preds = gender_net.forward()
#         print(gender_preds)
        gender = gender_list[gender_preds[0].argmax()]
        print(gender)
        male = 0
        female=0
        if gender =='Male':
            male=male+1
        elif gender == 'Female':
            female = female+1
    
        cv2.putText(img, gender, (x,y-5), cv2.FONT_HERSHEY_PLAIN,2,(0,0,250),2)
    
    cv2.imshow("Gender", img)
    if cv2.waitKey(1)==ord('q'):
        break
print("male: ", male)
print("female: ", female)
cam.release()
cv2.destroyAllWindows()
    
```
